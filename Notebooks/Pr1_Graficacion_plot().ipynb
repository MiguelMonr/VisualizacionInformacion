## Práctica 1. Graficación en R - plot()

R cuenta con varios paquetes gráficos. En este curso utilizaremos principalmente `ggplot2` y `leaflet` aunque es casi inevitable conocer y utilizar algunas de las funciones que ofrece el paquete básico `graphics`.  Eso es lo que haremos en esta práctica.

La función `plot()` permite crear gráficos de puntos dispersos (*scatterplot*), líneas y muchos otros tipos de gráficos, dependiendo de los argumentos que reciba.

Su sintaxis general es:

`plot(x,y,type,main,xlab,ylab,pch,col,las,…)`

| Parámetro  | Descripción                                                  |
| ---------- | ------------------------------------------------------------ |
| x, y       | Típicamente, vectores de valores correspondientes a las variables X y Y. Deben ser de la misma longitud. |
| type       | El tipo de gráfico a dibujar (puntos, líneas, etc.)          |
| main       | Título del gráfico                                           |
| xlab, ylab | Etiquetas para los ejes X y Y                                |
| pch        | Forma de los puntos                                          |
| col        | Color de símbolos y líneas                                   |
| las        | Estilo de las etiquetas                                      |
| …          | Muchos otros parámetros opcionales                           |

### 1. Creando un primer gráfico
Vamos a graficar la función coseno de menos PI a PI sin ningún parámetro adicional:

```r
x<- seq(-pi,pi,0.1)
y<-cos(x)
plot(x,y)
```

### 2. Tipos de gráficos
Por omisión se crea una gráfica con puntos como marcadores. Otros marcadores y otros tipos de graficado se establecen con el argumento `type`. Estos son algunos de los tipos disponibles:

| Valor | Descripción                       | Valor | Descripción                   |
| :---: | --------------------------------- | :---: | ----------------------------- |
|  "p"  | Puntos                            |  "l"  | Líneas                        |
|  "b"  | Puntos y líneas                   |  "o"  | Puntos y líneas sobreescritas |
|  "h"  | Líneas verticales tipo histograma | "s/S" | Gráfico en escalón            |

**PREGUNTA 2.** Grafique la función con una línea

```r
plot(x, y, type= "l")
```


Este snippet muestra algunos de los tipos de gráfico disponibles en R:

```r
# Matriz de 2x3 gráficos
par(mfrow = c(2, 3)) #! Aqui especificas las dimensiones de la matriz

plot(x, y, type = "p", main = "type = 'p'")
plot(x, y, type = "b", main = "type = 'b'")
plot(x, y, type = "o", main = "type = 'o'")
plot(x, y, type = "s", main = "type = 's'")
plot(x, y, type = "S", main = "type = 'S'")
plot(x, y, type = "h", main = "type = 'h'")
```



Para las líneas, podemos especificar el tipo de línea con el argumento `lty=n`, donde `n = [1,2, ..., 6]` y su grosor el argumento `lwd=m`.

El tamaño de los puntos se puede modificar con el argumento `cex`(*character expansion*). En la siguiente figura se muestra un gráfico de un segmento de la función con el tamaño por default y a la derecha, con una expansión de 1.5:


```r
#! Pregunta 3
par(mfrow = c(1, 2))

plot(x, y, pch= 8, type="o")
plot(x, y, pch=23, type= "o", col= "blue", bg="lightblue")
```

```r
par(mfrow = c(1, 2))

plot(x, y, pch=5)
plot(x, y, pch=5, cex=1.5) #! cex se usa para determinar el tamaño de la grafica
```

### 5. Añadir líneas, puntos, líneas de referencia y leyendas

Podemos añadir otras curvas a nuestra gráfica con las funciones `lines()` y `points()` que se invocan después de la función `plot()`:

```r
#! Pregunta 6
plot(x, y, pch = 8, type = "o", main= "grafica para el final",las= 1, xlab= "prueba1", ylab= "prueba2")

# TODO: Lo que cambia con el argumento las es la orientacion de la escala
```

```r
#! Pregunta 7
plot(x, y, type = "l", main = "Gráficas superpuestas", xlab = "X en [-Pi, Pi]", las = 1)
legend("topleft", legend = c("cos(x)", "sen(x)", "cos(x)/2"), fill = c("black", "red", "blue"))
grid(nx = 10, ny = 10, col = "black", lty = "dashed", lwd = 0.5)
lines(x, sin(x), type = "l", col = "red")
points(x, cos(x) / 2, type = "l", col = "blue")
```

```r
plot(x,y, type='l', main= 'Gráficas superpuestas', xlab='X en [-Pi, Pi]',las=1)
lines(x,sin(x),type='l', col='red')
points(x,cos(x)/2, type='l', col='blue')
```

En estos casos, es necesario añadir una leyenda que indique qué grafica cada curva. Esto se hace con la función `legend()` que recibe como argumentos la posición donde aparecerá la leyenda, un vector con el texto a desplegar para cada curva en el orden en que se generaron las gráficas, y los tipos de símbolo y colores utilizados en cada una.  

Si no se utilizaron símbolos, con el argumento `fill` se pueden especificar los colores de cada curva, que se distinguirán por un pequeño cuadrado del color correspondiente.

### 8. Guardar una gráfica

Para guardar una gráfica como una imagen en un archivo, debemos seguir tres pasos en secuencia:

1. Invocar una función para abrir un archivo de imagen, como `png()`, `jpg()` o `pdf()`.
2. Invocar la función `plot()` para generar la gráfica.
3. Invocar la función `dev.off()` para cerrar el archivo de imagen.

En el siguiente script, guardaremos la gráfica que hicimos en la sección anterior, como una imagen jpg



```r
jpeg(filename="demoImg.jpg", width=648, height=432)
plot(iris$Petal.Length, iris$Petal.Width, col=iris$Species, pch=19)
dev.off()
```
