Intento para hacer grafica de barras apiladas

```r
library(ggplot2)
library(dplyr)

setwd("/Users/miguela.monreal/CodigosVisInfo/codigos/")
data <- read.csv("v2Abril.csv")
head(data, 6)


summary(data)
```

```r
#Creando Dataframes para valores del 2004

datos_2004 <- data %>%
    filter(Year == 2004)
head(datos_2004)
```

```r

ggplot(datos_2004, aes(x = Region, y = Soda, fill = State.Abbr.)) +
    geom_bar(stat = "identity")+geom_hline(yintercept= 0, color = "red", linetype = "dashed", linewidth= 1.5) +
    labs(x = "Región", y = "Soda", fill = "Estado") +
    ggtitle("Consumo de Soda por Región en 2004") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1)) # Rotar etiquetas en el eje X si es necesario
```
