# Aclaracion!: 
En el metodo `geom_sf`si puedes elegir que dataframe graficar si lo que quieres es usar varias capas

```r
datos <- ozmap_states
ggplot(datos) +
    geom_sf(mapping = aes(fill = NAME)) + #* geom_sf es el metodo para graficar. En mapping ponemos las variables a corolear
    coord_sf() #* Necesario para indicar las coordenadas que se estan utilizando
```

```r
distritos <- rmapshaper::ms_simplify(ozmaps::abs_ced)

ggplot(datos) +
    geom_sf(mapping = aes(fill = NAME), show.legend = F) +
    geom_sf(data = distritos, fill = NA) +
    geom_sf_label(data = datos, aes(label = NAME), label.size = 0.025) +
    coord_sf()
```
