## La libreria que se usa es
`ggplot()`

ggplot funciona por capas. En este caso el orden de las capas si altera el resultado final 

```r
#Este es un ejemplo
library(ggplot2)
ggplot(data = economics, aes(x = date, y = unemploy)) + #*Aqui en el + es donde se ponen las apas
    geom_point()
```

Metodos importantes al trabajar con graficas: 
- Para temas de imagen usas la funcion `aes()`
- En `aes()` pones las columnas que quieres graficar. 
Dentro de la funcion de aes puedes elegir el color de la linea y el relleno
- Tipo de linea: `type=`
- Para elegir el color del relleno: `fill= `
- Para elegir el color de la linea: `col= `

```r
miGraf <- ggplot(data = mtcars, aes(x = wt, y = mpg))

miGraf +
    geom_point(size = 2, color = "red") +
    geom_smooth(method = "lm", se = FALSE)
```

## IMPORTANTE

Si bien `fill`y `col` tienen la misma funcion depende mucho el tipo de grafica el parametro que vamos utilizar.

- Col se usa en las siguientes graficas: scatter plots, lineas
- fill se usa para establecer el color del interior de los objetos que tienen un rea rellenable: barras, boxplots, densidad

# Heatmaps
*IMPORTANTE* Para heatmaps los datos tienen que estar en formato wide&long

```r
install.packages("reshape2")
```

```r
library(reshape2)
library(tidyr)

# agregamos col. Nombre auto
mtdf <- melt(mtcars) #! Este es importante 
mtdf$car <- rep(row.names(mtcars), 11)
head(mtdf)


# Tenemos que rescalar (0 a 1) porque los valores de un atributo (disp) son
#  mucho mayors que los dems
library(plyr)
library(dplyr)
library(scales)
mtdf <- ddply(mtdf, .(variable), transform, rescale = rescale(value))
# crea mapa
miGraf <- ggplot(mtdf, aes(variable, car))
miGraf + geom_tile(aes(fill = rescale), colour = "white") +
    scale_fill_gradient(low = "white", high = "steelblue")
```

# Grafica de densidad

```r
## Observa que podemos solicitar que calcule densidad en vez de conteo
ggplot(iris, aes(x = Sepal.Width)) +
    geom_histogram(aes(y = ..density..), bins = 10) + #* Se utiliza el parametro ..density..
    geom_density(adjust = 2) #* Usas el parametro geom_density
# geom_density(adjust=2,kernel='rectangular')
```

Graficas al 100%

```r
# Propocion de cada elemento a la totalidad de dataset
ggplot(mtcars, aes(x = cyl, fill = factor(gear))) +
    geom_bar(position = "fill") +
    stat_count(
        geom = "text",
        aes(label = paste(round((..count..) / sum(..count..) * 100), "%")),
        position = position_fill(vjust = 0.5), colour = "white"
    )
```

Con proporciones

```r
# Propocion de cada elemento a la totalidad de la columna
# Calcula los porcentajes
mtcars_summary <- mtcars %>%
    group_by(cyl, am) %>%
    summarise(count = n()) %>%
    mutate(percent = count / sum(count) * 100)

# Crea el grfico de barras apiladas al 100% con porcentajes
ggplot(mtcars_summary, aes(x = cyl, y = percent, fill = factor(am))) +
    geom_bar(stat = "identity", position = "fill") +
    geom_text(aes(label = paste0(round(percent), "%")),
        position = position_fill(vjust = 0.5), color = "white"
    )
```
